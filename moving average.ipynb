```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import d2l.torch as d2l

d2l.use_svg_display()

# 使用你提供的数据
x = np.arange(10000)
y = np.random.normal(0, 1, 10000) + np.log(x+1e-1)
z = y/2

data = {'steps': x, 'rewards': y}
df = pd.DataFrame(data)

window_size = 50  
df['rewards_moving_avg'] = df['rewards'].rolling(window=window_size).mean()

plt.figure(figsize=(10, 6))

plt.plot(df['steps'], df['rewards'], color='lightgray', label='Reward')
plt.plot(df['steps'], df['rewards_moving_avg'], color='blue', label='Moving Average Reward')
plt.title('Reward over Steps')
plt.xlabel('Steps')
plt.ylabel('Reward')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import d2l.torch as d2l

d2l.use_svg_display()

# 使用你提供的数据
x = np.arange(10000)
y = np.random.normal(0, 1, 10000) + np.log(x+1e-1)
z = (np.random.normal(0, 1, 10000) + np.log(x+1e-1))/2

data = {'steps': x, 'rewards': y, 'z_values': z}
df = pd.DataFrame(data)

window_size = 50
df['rewards_moving_avg'] = df['rewards'].rolling(window=window_size).mean()
df['z_values_moving_avg'] = df['z_values'].rolling(window=window_size).mean()

plt.figure(figsize=(10, 6))

plt.plot(df['steps'], df['rewards'], color='lightgray', label='Reward')
plt.plot(df['steps'], df['rewards_moving_avg'], color='blue', label='Moving Average Reward')

# 添加x和z的数据
plt.plot(df['steps'], df['z_values'], color='lightgreen', label='Z Value')
plt.plot(df['steps'], df['z_values_moving_avg'], color='green', label='Moving Average Z Value')

plt.title('Reward and Z Value over Steps')
plt.xlabel('Steps')
plt.ylabel('Value')
plt.legend()
plt.show()
```
